Nous allons tenter de comprendre les élèments qui composent la syntaxe d'une boucle.

Pour cela, nous utiliserons la fonction `print` dont voici quelques exemples :

```python
print(5)
```

```python
toto = 5
print(toto)  # remarquez l'absence de guillements...
```

Passons à présent à notre 1ere boucle `for indice in range(4):`

```python
for indice in range(4):
```

Cela commence mal... Python nous dit qu'il manque quelque chose...

```python
for indice in range(4):
    toto = 5            #remarquez l'indentation, comme pour les fonctions.
```

Là, plus de problème, mais il ne se passe pas grand chose... Utilisons la fonction `print`.

```python
for indice in range(4):
    toto = 5 
    print(toto)
```

```python
toto
```

Visiblement, cette boucle peut nous servir à répéter plusieurs fois la même instruction, sans devoir la taper à chaque fois.

Essayons autre chose :

```python
for indice in range(4):
    toto = 5            
    print(indice)
```

```python
toto
```

la variable `toto` contient bien 5 et visiblement la variable `indice` a pris les valeurs 0,1,2 et 3.

Parlons à présent de "factorielle" :


Selon Wikipédia :


<div style = "background-color:#eeeeee;">
    
En mathématiques, la factorielle d'un entier naturel $n$ est le produit des nombres entiers strictement positifs inférieurs ou égaux à $n$.

Cette opération est notée avec un point d'exclamation, $n!$, ce qui se lit soit « factorielle de $n$ », soit « factorielle $n$ » soit « $n$ factorielle » (cette dernière expression est la moins utilisée). Cette notation a été introduite en 1808 par Christian Kramp. 
    
</div>

Par exemple : 

$5! = 1\times2\times3\times4\times5=120$

Par convention : $0! = 1$.

```python
from random import randint  #du module qui se nomme random on importe la fonction randint
def lancer():
    resultat = randint(1,6)
    return resultat
    
print(lancer())
```
